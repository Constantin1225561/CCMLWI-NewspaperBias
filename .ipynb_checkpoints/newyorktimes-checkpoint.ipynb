```python
from nytimesarchive import ArchiveAPI
import pandas as pd
```

```python
api = ArchiveAPI('6faa534a59e44205829aad054e5095ce')
years = range(2010,2017)
#jan = api.query(2017,4)
```

```python
docs = []
for y in years:
    for m in range(1,13):
        print("Querying {},{}".format(m,y))
        docs.append(api.query(y,m)['response']['docs'])
```

```python
len(docs)
```

```python
def is_about_refugees(doc):
    for keyword in doc['keywords']:
        if (keyword['name']=='subject' and 'refugee' in keyword['value'].lower()):
            return True
```

```python
refugee_docs = [doc for doc in docs if is_about_refugees(doc)]
```

```python
len(refugee_docs)
```

```python
df = pd.DataFrame.from_dict(refugee_docs)
```

```python
def extract_main_headline(headline):
    return headline['main']
```

```python
df['headline'] = df['headline'].apply(extract_main_headline)
```

```python
headlines = df[['pub_date','web_url','headline']].copy()
```

```python
headlines
```

```python
headlines.to_csv('headlines_nyt.csv',index=False)
```
